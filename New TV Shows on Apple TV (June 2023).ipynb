```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.rottentomatoes.com/browse/tv_series_browse/affiliates:apple_tv~audience:upright~critics:fresh~sort:newest"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
tv_series = soup.find_all("div", class_="discovery-tiles__wrap")

data_judul = []
data_eps = []

for series in tv_series:
    juduls = [span.get_text(strip=True) for span in series.find_all('span', class_='p--small', attrs={'data-qa': 'discovery-media-list-item-title'})]
    eps = [span.get_text(strip=True) for span in series.find_all('span', class_='smaller', attrs={'data-qa': 'discovery-media-list-item-start-date'})]
    if len(juduls) == len(eps):
        data_judul.extend(juduls)
        data_eps.extend(eps)

pd.DataFrame({"Title": data_judul, "Start Date": data_eps}).to_excel('New TV Shows on Apple TV (June 2023)
.xlsx')

```
